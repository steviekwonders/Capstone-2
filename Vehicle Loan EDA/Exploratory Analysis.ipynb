```python
# Function for plotting a histogram.
def plothistogram(var_name, bins, data):
    plt.subplots(1,2,figsize=(12,4))
    plt.subplot(1,2,1)
    ax = data[var_name].hist(bins = bins, color = 'b')
    ax.set_ylabel('Count') 
    ax.set(xlim=(data[var_name].quantile(0.01), data[var_name].quantile(0.99)))   
    ax.set_xlabel(var_name)
    ax.set_title('Overall Frequency distribution for ' + var_name)
    print('Summary statistics for', var_name)
    return data[var_name].describe()
```

```python
def createviolinplot(var_name, data):
    # This code block will create a violin plot for continuous variables. It will make two violin plots for defaulters 
    # and non-defaulters ('loan_default' value of 1 and 0 respectively). It will also return the t-test statistic in
    # difference of means to verify if the two samples have the same average.
    plt.subplot(1,2,2)
    sns.violinplot(x = 'loan_default', y=var_name, data= data)
    defaulter = data[data['loan_default']==1][var_name]
    nondefaulter = data[data['loan_default']==0][var_name]
    print('Nondefaulter mean for', var_name, ':' ,nondefaulter.mean(), ', std:', nondefaulter.std()) 
    print('Defaulter mean for ', var_name, ': ', defaulter.mean(), ', std:', defaulter.std())
    return scipy.stats.ttest_ind(defaulter, nondefaulter, equal_var = False)
```

```python
# Function to plot barplot for nondefaulter. Additionally, return the chi-square test stat between default (obs_freq) and non-default (exp_freq)
def plotbarnondefaulter(var_name,data):
    # This code block will create a barplot for categorical variable.
    defaulter = data[data['loan_default']==1]
    nondefaulter = data[data['loan_default']==0]
    
    print('Blue bar plot is the NonDefaulter bar plot for ' + var_name)
    nondefaulter.groupby([var_name])[[var_name]].count().plot(kind='bar', color='b').set_ylabel('Non-defaulter Count')

    
    # This code block will create two dictionaries for defaulters and nondefaulters for specified variable. It will obtain
    # percentage for each categorical value which we can use in chi-square test to see if there is difference in their distribution.
    var_name_values = list(df[var_name].unique())
    exp_freq = {}; obs_freq = {}
    
    for val in var_name_values:
        if val not in list(defaulter[var_name].unique()):
            obs_freq[val] = 0
        else:
            obs_freq[val] = defaulter[defaulter[var_name] == val][var_name].count()/len(defaulter)
            
    for val in var_name_values:
        if val not in list(nondefaulter[var_name].unique()):
            exp_freq[val] = 0
        else:
            exp_freq[val] = nondefaulter[nondefaulter[var_name] == val][var_name].count()/len(nondefaulter)
            
    #print('expected', exp_freq)
    #print('observed' ,obs_freq)
    
    return print('Chi-square test results:',scipy.stats.chisquare(list(obs_freq.values()), list(exp_freq.values()) ) )            # do chi-square on dictionary values
```

```python
def plotbardefaulter(var_name, data):
    # This code block will create a bar plot of specified categorical variable for defaulters. This will be used
    # if chi-square shows significance and we want to see how the distribution differs from the overall barplot. 
    defaulter = data[data['loan_default']==1]
    # print('Red bar plot is the Defaulter bar plot for ' + var_name)
    defaulter.groupby([var_name])[[var_name]].count().plot(kind='bar', color='r').set_ylabel('Defaulter Count')
    return print('Red bar plot is the Defaulter bar plot for ' + var_name)
```
