```python
# Save the feature names into a list: features
feature = list(df.columns)
```

```python
# Save features of type 'int'. 'categ_features' 
categ_features = []
# Save features of type 'float' into num_features.
num_features = []

# Assume, for now, that if feature is of type int, it's categorical and add to categorial list.
for val in feature:
    if df[val].dtypes == 'int64' or df[val].dtypes == 'object':
        categ_features.append(val)
    else:
        num_features.append(val)

print('Categorical features: ', categ_features); print('\n')
print('Numerical features: ', num_features)
```

```python
# Assuming features of type int to be categorical was a bad idea. Remove features in categorial list that should be 
# numerical (should be plotted via histogram) and add it to numerical features list.
to_be_numerical = ['disbursed_amount', 'asset_cost', 'PERFORM_CNS.SCORE', 'PRI.NO.OF.ACCTS', 
                   'PRI.ACTIVE.ACCTS', 'PRI.OVERDUE.ACCTS', 'PRI.CURRENT.BALANCE', 'PRI.SANCTIONED.AMOUNT', 'PRI.DISBURSED.AMOUNT',
                   'SEC.NO.OF.ACCTS', 'SEC.ACTIVE.ACCTS', 'SEC.OVERDUE.ACCTS', 'SEC.CURRENT.BALANCE', 'SEC.SANCTIONED.AMOUNT', 
                   'SEC.DISBURSED.AMOUNT', 'PRIMARY.INSTAL.AMT', 'SEC.INSTAL.AMT', 'NEW.ACCTS.IN.LAST.SIX.MONTHS', 
                   'DELINQUENT.ACCTS.IN.LAST.SIX.MONTHS', 'NO.OF_INQUIRIES']

for feat in to_be_numerical:
    categ_features.remove(feat)
    num_features.append(feat)
    
print('Categorical features ', categ_features, '(length): ',  len(categ_features)); print('\n')
print('Numerical features  ', num_features, '(length): ', len(num_features)); print('\n')
print('Shared feature(s) between categorical and numerical lists:', list(set(categ_features) & set(num_features)))
```

```python
# Function for plotting a histogram.
def plothistogram(var_name, bins):
    plt.subplots(1,2,figsize=(12,4))
    plt.subplot(1,2,1)
    ax = df[var_name].hist(bins = bins, color = 'b')
    ax.set_ylabel('Count') 
    ax.set(xlim=(df[var_name].quantile(0.01), df[var_name].quantile(0.99)))   
    ax.set_xlabel(var_name)
    ax.set_title('Overall Frequency distribution for ' + var_name)
    return df[var_name].describe()
```

```python
def plotdefaulter(var_name, bins):
    plt.subplot(1,2,2)
    defaulter = df[df['loan_default']==1]
    ax = defaulter[var_name].hist(bins = bins, color = 'r')
    ax.set_ylabel('Count') 
    ax.set(xlim=(defaulter[var_name].quantile(0.01), defaulter[var_name].quantile(0.99)))   
    ax.set_xlabel(var_name)
    ax.set_title('Defaulter Frequency distribution for ' + var_name)
    return defaulter[var_name].describe()
```

```python
# Function to plot barplot. 
def plotbarplot(var_name):
    df.groupby([var_name])[[var_name]].count().plot(kind='bar', color='b').set_ylabel('Count')
```

```python
def plotbardefaulter(var_name):
    defaulter = df[df['loan_default']==1]
    defaulter.groupby([var_name])[[var_name]].count().plot(kind='bar', color='r').set_ylabel('Default Count')
```
